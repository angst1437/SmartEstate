```python
class LogbookParser:
    def __init__(self, url):
        chrome_version = 135
        options = uc.ChromeOptions()
        options.add_argument('--disable-blink-features=AutomationControlled')
        options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.6998.89 Safari/537.36')
        
        self.browser = uc.Chrome(options=options)
        self.url = url
        
        
    def loading_ads(self, browser):
        browser.get(self.url)
        wait = WebDriverWait(browser, 10)

        ad_links = set()
        attempts = 0

        while attempts < 50:
            soup = BeautifulSoup(browser.page_source, 'html.parser')
            a_tags = soup.find_all('a', href=True)
            last_card = None

            # Находим последний элемент-карточку 
            for a_tag in reversed(a_tags):
                if a_tag.find('figure', attrs={"data-test-component": "ProductCard"}):
                    last_card = a_tag['href']
                    break

            browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            try:
                # Ждём появления нового элемента 
                wait.until(lambda d: last_card not in [a.get_attribute('href') for a in d.find_elements(By.CSS_SELECTOR, 'a[href]')])
            except:
                attempts += 1
                continue

            soup = BeautifulSoup(browser.page_source, 'html.parser')
            a_tags = soup.find_all('a', href=True)

            for a_tag in a_tags:
                if (a_tag.find('figure', attrs={"data-test-component": "ProductCard"}) and
                    '/nedvijimost/' in a_tag['href'] and
                    not a_tag['href'].startswith('https://click.youla.io')):

                    href = a_tag['href']
                    if href.startswith('/'):
                        full_url = f"https://youla.ru{href.split('?')[0]}"
                        ad_links.add(full_url)

            print(f"Прокрутка {attempts + 1}: всего уникальных ссылок {len(ad_links)}")
            attempts = 0 

        return list(ad_links)

    def parse_urls(self):
        ad_links = self.loading_ads(self.browser)  

        for idx, url in enumerate(sorted(ad_links), 1):
            print(f'{idx}. {url}')

        return ad_links

        
```

```python
if __name__ == '__main__':
    parser = LogbookParser('https://youla.ru/all/nedvijimost')
    ads = parser.parse_urls()
    print(f"Найдено {len(ads)} объявлений")
```
